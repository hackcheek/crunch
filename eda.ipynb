```python
import pandas as pd
import glob
import os
import json
```

```python
def make_rows(paths):
    for path in paths:
        with open(path, 'r') as f:
            data = json.load(f)
        
        for d in data:  
            d.pop('prevObject')
            d.pop('length')
            row = {}
            for v in d.values():
                row.update(v)
            yield row

def build_dataframe(dirname='companies_data'):
    data_paths = glob.glob(os.path.join(dirname, '*.json'))
    return pd.DataFrame(make_rows(data_paths))
    
```

```python
df = build_dataframe()
df.drop(columns=['last_key_employee_change_date', 'growth_insight_indicator', 'last_funding_at', 'last_funding_type'])
```
